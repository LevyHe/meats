# Plotting a pie chart of value counts



https://stackoverflow.com/questions/36963502/cant-plot-value-counts-for-pie-chart

```python
%matplotlib inline
import pandas as pd
a = pd.DataFrame({'a': [1,0,0,0,1,1,0,0,1,0,1,1,1],'b': [1,0,0,0,1,1,0,0,1,0,0,0,0]})
a.apply(pd.value_counts).plot.pie(subplots=True)
```

```python
print(a.a.value_counts())
a.a.value_counts(True)
```

```python
print(a.b.value_counts())
a.b.value_counts(True)
```
