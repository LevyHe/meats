```python
%matplotlib notebook

# see more:
# https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html
    
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
```

```python
x, y = np.mgrid[:10, :10]
z = x - y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# ax.plot_surface(x, y, z) - this fails,
# coordinates need to be flattened.

ax.scatter(x.flatten(), y.flatten(), z.flatten())
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
